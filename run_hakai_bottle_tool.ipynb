```python
# Import the packages needed
from hakai_bottle_tool import hakai_bottle_tool
import plotly.express as px
import pandas as pd

# Hakai will ask for your Hakai Credentials the first time
```

# Retrieve Bottle Data for a station and time range
Select specific Hakai Station and time range

```python
# Give an example to generate bottle data from QU39 for the years 2019 to 2021.
station='QU39'
time_min='2014-01-01' # YYYY-MM-DD
time_max='2021-01-01' # YYYY-MM-DD
```

```python
# Retrieve the bottle data
df_bottles = hakai_bottle_tool.get_bottle_data(station=station,time_min=time_min,time_max=time_max)

# Show me the resulting bottle data
df_bottles
```

# Review Matching Result

```python
# Plot unmatched data
df_bottles['unmatched'] = False
df_bottles.loc[df_bottles['ctd_hakai_id'].isna(),'unmatched'] = True
fig = px.scatter(df_bottles,x='collected',y='depth',color='unmatched')
fig.update_yaxes(autorange='reversed')
fig.show()
```

```python
# Bottle distribution
fig = px.histogram(df_bottles,x='depth')
fig.update_traces(xbins={'start':0,'end':df_bottles['depth'].max(),'size':1})
```

# Example Plots
Here's some example plots which are using the [plotly package](https://plotly.com/python/).

```python
# Show an example plot of the resulting data
fig = px.scatter(df_bottles,x='time',y='depth',color='ctd_dissolved_oxygen_ml_l',hover_data=['nutrients_hakai_id'])
fig.update_yaxes(autorange='reversed')
fig.show()
```

```python
# Or let's compare CTD vs Nutrient_PO4
# Show an example plot of the resulting data
fig = px.scatter(df_bottles,x='ctd_salinity',y='nutrients_po4',color='ctd_dissolved_oxygen_ml_l',range_color=[2,8],color_continuous_scale='oxy')
fig.show()
```

# Filter the variables to necessary ones
Generating those bottle files generates a lot of different variables (>600) due to the different aggregating methods (mean, joined strings, count, |max-min|).

A predefined list of variables is available to help simplify the list of the variables available.

```python
df_bottles = hakai_bottle_tool.filter_bottle_variables(df_bottles,'Reduced')
```

# Generate File from the outputed data
You can save the resulting data to NetCDF or CSV. Data is saved by default within the active directory. 


---


***If run within Google Colab, click on the folder symbol*** (image.png) ***along the left side of the browser to access the directory.***

```python
# As NetCDF
hakai_bottle_tool.export_to_netcdf(df_bottles)

# or as CSV
df_bottles.to_csv('Hakai_Bottle_Data_QU39_20190101_20210101.csv')
```
